```
# Carregando os pacotes necessários 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn  as sns 
```

```
# Carregando o banco de dados PNAD
pnad = pd.read_csv('https://raw.githubusercontent.com/neylsoncrepalde/introducao_ao_python/master/pes_2012.csv')
```

```
pnad["V0302"].value_counts()  # Mostrando a quantidade de pessoas por sexo
```

```
# Gráficos - Sexo (V0302)

pnad.V0302.head()
pnad["sexo"] = pnad.V0302.astype("category") # renomeando a variável

sns.countplot(x=pnad.sexo)
plt.title(         "186.054                                   176.397")
plt.xlabel("Sexo")
plt.ylabel("Distribuição de sexo no Brasil - 2012")
plt.show()

# No banco estudado, existem mais mulheres do que homens
```

```
# Quantidade de pessoas por Cor/Raça
pnad.raca.value_counts()
```

```
# Gráficos - Raça (V0404)

pnad["raca"] = pnad.V0404 # renomeando a variável

sns.set(color_codes = True)
sns.countplot(y=pnad.raca)
plt.title("Distribuição de Cor/Raça no Brasil - 2012")
plt.ylabel("Cor/Raça")
plt.xlabel("Frequência")
plt.show()

# A raça mais frequente no banco é a parda, seguida de branca e preta
```

```
pnad.UF = pnad.UF.astype("category")
sns.countplot(y=pnad.UF)
sns.set_style("ticks", {"xtick.major.size": 5, "ytick.major.size": 5})
plt.title("Pessoas por Estados")
plt.xlabel("Quantidade de Pessoas")
plt.ylabel("Pessoas por Estado")
```

```
# Gráfico - Renda por idade (V8005) - scatter

pnad["idade"] = pnad.V8005 # renomeando a variável

plt.scatter(x = pnad.renda, y = pnad.idade, color="#F08080")
plt.title("Renda por Idade")
plt.ylabel("Idade")
plt.xlabel("Renda")
plt.show()
```

```
# Gráfico - Idade (V8005) por Cor/Raça (V0404)

pnad["cor"] = pnad.V0404 # renomeando a variavel
pnad["idade"] = pnad.V8005 # renomeando a variavel

sns.boxplot(y = 'cor', x = 'idade', data = pnad)
plt.title("Boxplot das variaveis cor e idade")
plt.ylabel("Cor")
plt.xlabel("Idade")
plt.show()
```
